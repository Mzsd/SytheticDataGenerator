```python
import pgeocode
import pandas as pd
from fng_api import *
```

```python
gb = pgeocode.Nominatim('gb')
# gb.query_postal_code("G34 0EG")
```

```python
postcodes_df = pd.read_csv("smalluser-glasgow.csv")
```

```python
postcodes_df
```

```python
def fetch_postcode_data(postcode):
    try:
        data = gb.query_postal_code(postcode)
        return f"{data.latitude}, {data.longitude}"
    except:
        return None
```

```python
postcodes_df['latlong'] = postcodes_df['Postcode'].apply(lambda x: fetch_postcode_data(x))
```

```python
postcodes_df[['latitude', 'longitude']] = postcodes_df['latlong'].str.split(',', expand=True)
```

```python
final_postcodes_df = postcodes_df[
    ['Postcode', 'latitude', 'longitude']
]

final_postcodes_df.columns = ['postcode', 'latitude', 'longitude']
```

```python
final_postcodes_df.to_csv("postcode_latlong.csv", index=False)
```
